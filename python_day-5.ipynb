## while

```python
i=2
while i<11:    # or i<=10
    print(i)
    i+=1
```

```python
i=5
while i>0:
    print(i)
    i-=1
```

```python
# print the multipilcation table
num=int(input('enter the number'))
i=1
while i<11:
    print(num,'*',i,'=',num*i)
    i+=1
```

## else

```python
i=2
while i<11:    # or i<=10
    print(i)
    i+=1
else:
    print('finished')
```

## Break,Continue,Pass

```python
for i in range(2,10):
    if i==7:
        break
    print(i)
```

```python
for i in range(2,10):
    if i==7:
        continue
    print(i)
```

```python
for i in range(6,100):
    pass
```

### nested loop

```python
for i in range(1,5):
    for j in range(1,5):
        print(i,j)
```

```python
n=int(input())
for i in range(1,n):
    for j in range(1,i+1):
        print('*',end='')
    print()
```

```python
# i like python
# vowel *
# * l*k* pyth*n
text="i like python"
vowels="aeiouAEIOU"
result=""
for char in text:
    if char in vowels:
        result+="*"  
    else:
        result+=char
print(result)
```
