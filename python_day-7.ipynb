```python
# set
s={}        # empty set is called dictionary
type(s)     # if we add any thing into it,it becomes set
```

```python
s=set()
type(s)
```

```python
l={9,3,2,0,'p',999999,9.8,9,85605587977777777}        #set is unordered
l
```

```python
s={1,2,3,4,5,6,7,7,8,7,7}                # set removes duplicate elements
s
```

```python
l=[9,8,9,9,0]
l1=set(l)
l1
```

```python
s={1,2,3,(4,6,7)}   # we can only store immutubale(tuple) in the set ,we cannot store list or mutable item in the set
s                   # we cannot create nested set also
```

```python
s1={1,2,3}
s2={4,5,6}
s1.union(s2)
```

```python
s1={1,2,3}
s2={4,5,6}          # | is union sign
s1|s2
```

```python
s1={1,2,3,4,5}
s2={4,5,6}
s1&s2                    # & is intersection sign
```

```python
s1={1,2,3,4,5}
s2={4,5,6}
s1-s2                      # - is used for difference
```

```python
s1={1,2,3,4,5,7}
s2={4,5,6,7,8}             # opposite for intersection
s1^s2                     # ^ is used for symmetric difference
```

```python
s={5,7,9}
s.add(50)             # add is used to add one value
s
```

```python
s={5,7,9}
s.update({6,8,90})      # update is used to add multiple values
s
```

```python
s={5,7,9}
s.update((6,8,90))
s
```

```python
s={4,6,7,90}
s.remove(7)           # remove helps to remove any element in the set
s
```

```python
s={4,6,7,90}
s.discard(900)          # discard doesnot throw any error even the element is not available in the set
s 
```

```python
s={4,6,7,90}
s.pop()             # pop deletes any item randomly in set
s
```

```python
s={4,6,7,90}
s.clear()         # clear removes all the elements and give empty set
s
```

```python
s1={1,2,3,4,5}
s2={3,4,5}
s1.issuperset(s2)     # super set is true if all the elements of s2 are present in s1
```
