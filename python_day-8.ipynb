```python
# The number is even or odd
# with argunment with return
def even_odd(n):
    if n%2==0:
        return 'even'
    else:
        return 'odd'
even_odd(40)
```

```python
# no arg but with return
def even_odd():
    n=int(input())
    if n%2==0:
        return 'even'
    else:
        return 'odd'
even_odd()
```

```python
def palindrome():
    n=input()
    if n==n[::-1]:
        return 'palindrome'
    else:
        return 'not palindrome'
palindrome()
```

```python
def palindrome(n):
    if n==n[::-1]:
        return 'palindrome'
    else:
        return 'not palindrome'
palindrome('dad')
```

```python
def pat(n):
    for i in range(n):
        for j in range(2**i):
            print('a',end='')
        for j in range(i+1):
            print('b',end='')
        print()
pat(4)
```

```python
# 1
# 121
# 12321
# 1234321
# 123454321
def pattern(n):
    for i in range(1,n+1):
        num=int('1'*i)
        print(num**2)
pattern(5)
```

```python
def fact(n):
    f=1
    for i in range(1,n+1):
        f *=i
    return f
fact(5)
```

```python
def sum(n):
    s=0
    for i in range(1,n+1):
        s=s+i
    return s
sum(10)
```
