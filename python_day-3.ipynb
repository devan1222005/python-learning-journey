# isalnum / isalpha / isdigit / isidentifier

```python
a='python123' 
a.isalnum()     #Tells wheather the given string is alpha numerical or not   # works only on the integers,not on decimals
```

```python
a='python'
a.isalpha()     #Tells wheather the given string is alphabet or not    # if we have a space in between the letters it will consder it as non alphabet
```

```python
a='123'
a.isdigit()     #Tells wheather the given string is numerical or not  # it only takes intergers,decimals will not be considered
```

```python
a='_4'
a.isidentifier()  # Tells wheather the given value is identifier as not,all identifer rules will be applied here
```
