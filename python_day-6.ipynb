# insert/delete/remove

```python
l=[1,6,7,9]     # insert is used to add the element in exact indexing
l.insert(1,60)  # It does not replaces the older element,but adds new element,pushes older element to next indexed position
l
```

```python
l=[1,6,7,9]
l.insert(3,91)
l
```

```python
l=[1,6,7,91,9]
del l[1:3]   #del is used to delete the values in the indexing position
l
```

```python
l=[1,'p','q','r',89,90,45]
del l[1:4]      #delete remove entire list
l
```

```python
l=[1, 89, 90, 45]
l.remove(45)   # remove is used to delete any element without using indexing
l
```

# pop/remove

```python
l=[1, 89, 90]
l.pop()        # pop is used to remove last element in the list
l
```

```python
l=[7,8,9,'python']
l.pop()
l
```

```python
l=[7,8,9,'python']
l.clear()          # clear is used to delete all elements in the list,still list is available
l
```
