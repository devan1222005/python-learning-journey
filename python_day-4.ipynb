```python
#try to check the strings is palindrome or not
s=input('enter the string')
if s==s[::-1]:           #use negative indexing and compare both strings
    print('palindrome')
else:
    print('not palindrome')
```

```python
# check the number is postive or negative
n=int(input('enter the number'))
if n>0:
    print('postive')
elif n==0:
    print('zero')
elif n<0:
    print('negative')
else:
    print('nothing')
```

```python
# If the temparature is between 30 to 50 hot
# If the temparature is between 10 to 29 modearate
# If the temparature is between 1 to 9 cool
# If the temparature is less than 1 dont go anywhere
temp=eval(input('enter the temparature'))
if temp>=1 and temp<=9:
    print('cool')
elif temp>=10 and temp<=29:
    print('modarate')
elif temp>=30 and temp<=50:
    print('hot')         
else:
    print('dont go any where')
```

```python
# if the number is divisble by 5 and 7 ,print divisible by 5 and 7
# if not divisible by 5 and 7,print(learn maths)
n=int(input('enter the number'))
if n%5==0 and n%7==0:
    print('divisible by 5 and 7')
elif n%5==0 and n%7!=0:
    print('ok')
elif n%5!=0 and n%7==0:
    print('not ok')
else:
    print('learn maths')
```
