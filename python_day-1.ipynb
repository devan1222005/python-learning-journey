## Operator

```python
# Arithmatic operator (+,-,*,/,//,%,**)
4+6 # addition
9-1 # subtraction
5*7 # multiply
```

```python
7/2 # standard division (point number as output)
7//2 # floor division (quotient as output)
7%2 # modulo divion (Remainder as output)
```

```python
2**3 # exponential / power
```

```python
# Relational operator (>,<,<=,>=,!=,==)
# output will be True or False
5<4 #lessthan
4>5 #greaterthan
4<=4 #lessthanequalto
5>=4 #greaterthanequalto
6!=9 #not equal to
4==4 #equal to
```

```python
# Logical Operator (and,or,not)
0 and 0
0 and 1
1 and 0
1 and 1 # If both are true only output will be true
```

```python
0 or 0
0 or 1
1 or 0
1 or 1 # If any one is true then output will be true
```

```python
not 1
not 0 #opposite will be the output (true=false,false=true)
```

```python
# Assignment operator (+=,-=,*=,/=,//=,%=,**=)
a=2
#a=a+2
a%=6
a
```

```python
a=9
a=a+2
a
```

```python
# Membership Operator (in ,not in)
'p' in 'python'
'xr' in 'xerox'
```
